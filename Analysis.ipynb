```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("cleaned_flipkart_ac.csv")
```

```python
df.shape
```

```python
plt.figure(figsize=(25,10))

# Brand Distribution
plt.subplot(2,2,1)
df['Brand'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Top 10 AC Brands on Flipkart')
plt.xlabel('Brand')
plt.ylabel('Count')
```

```python
# Set simple style
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (8, 5)
```

```python
# Top 10 Brands
top_brands = df['Brand'].value_counts().head(10)
plt.bar(top_brands.index, top_brands.values, color='skyblue')
plt.title("Top 10 AC Brands")
plt.xlabel("Brand")
plt.ylabel("Number of Models")
plt.xticks(rotation=45)
plt.show()
```

```python
# Price Distribution
plt.hist(df["Price ()"], bins=30, color='lightgreen', edgecolor='black')
plt.title("Price Distribution of ACs")
plt.xlabel("Price ()")
plt.ylabel("Count")
plt.show()
```

```python
# Average Price by Energy Rating
avg_price = df.groupby("Energy_Rating (Star)")["Price ()"].mean().reset_index()
plt.bar(avg_price["Energy_Rating (Star)"], avg_price["Price ()"], color='orange')
plt.title("Average Price by Energy Rating (Star)")
plt.xlabel("Energy Rating (Star)")
plt.ylabel("Average Price ()")
plt.show()
```

```python
# Rating Distribution
plt.hist(df["Rating"], bins=10, color='violet', edgecolor='black')
plt.title("Customer Rating Distribution")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()
```

```python
# Price vs Rating (Scatter Plot)
plt.scatter(df["Rating"], df["Price ()"], color='salmon', alpha=0.6)
plt.title("Price vs Rating")
plt.xlabel("Rating")
plt.ylabel("Price ()")
plt.show()
```

```python
# Average Price by Capacity (Ton)
avg_cap = df.groupby("Capacity (Ton)")["Price ()"].mean().reset_index()
plt.bar(avg_cap["Capacity (Ton)"], avg_cap["Price ()"], color='lightblue')
plt.title("Average Price by Capacity (Ton)")
plt.xlabel("Capacity (Ton)")
plt.ylabel("Average Price ()")
plt.show()
```

```python
# median price by model year
year_price = df.groupby("Model_Year")["Price ()"].median().reset_index()
plt.plot(year_price["Model_Year"], year_price["Price ()"], marker='o', color='green')
plt.title("Median Price by Model Year")
plt.xlabel("Model Year")
plt.ylabel("Median Price ()")
plt.show()
```

```python
# Correlation Heatmap (for numeric columns)
plt.figure(figsize=(5,4))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap (Numeric Features)")
plt.show()
```
