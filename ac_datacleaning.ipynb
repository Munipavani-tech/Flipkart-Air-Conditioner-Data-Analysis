```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("flipkart_ac_originaldata.csv")
```

```python
na_values = ['N/A', 'NA', 'na', 'NaN', 'nan', 'NONE', 'None', 'null', 'Null', '', '-', '--', '?']
df.replace(na_values, np.nan, inplace=True)
```

```python
# Strip extra spaces in string columns
for col in df.select_dtypes(include=['object']).columns:
    df[col] = df[col].astype(str).str.strip()
    df[col].replace({'nan': np.nan, '': np.nan}, inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
for col in df.columns:
    # Remove commas, %, and ₹ symbols before numeric conversion
    if df[col].dtype == 'object':
        df[col] = df[col].str.replace(',', '', regex=False).str.replace('%', '', regex=False).str.replace('₹', '', regex=False)
        df[col] = pd.to_numeric(df[col], errors='ignore')
```

```python
# Handle missing values
for col in df.columns:
    if df[col].dtype == 'object':
        # Fill text columns with most common value
        if df[col].isna().sum() > 0:
            df[col].fillna(df[col].mode()[0], inplace=True)
    else:
        # Fill numeric columns with median
        if df[col].isna().sum() > 0:
            df[col].fillna(df[col].median(), inplace=True)
```

```python
# Convert Model_Year to integer 
df['Model_Year'] = pd.to_numeric(df['Model_Year'], errors='coerce').astype('Int64')
```

```python
df.to_csv("cleaned_flipkart_ac.csv", index=False)
print("Cleaned dataset saved as 'cleaned_flipkart_ac.csv'")
```
